```python
import pandas as pd
import phreeqpy.iphreeqc.phreeqc_dll as phreeqc_mod

# Replace with the correct path if it's not in the same directory as the notebook
dfile_path = r"C:\Users\janel\OneDrive\Documents\IU grad\Research\Water-chem Project\McK water paper\codefiles\GLAQ.csv"
    # path for WW data: "C:\Users\janel\OneDrive\Documents\IU grad\Research\Water-chem Project\WarrenWood data\datafiles\Appendix3_4066subset.csv"
    # or "C:\Users\janel\OneDrive\Documents\IU grad\Research\Water-chem Project\WarrenWood data\datafiles\WWdata_Appendix3.csv"

# PHREEQC database path
pqdb = r"C:\Users\janel\OneDrive\Documents\IU grad\Research\Water-chem Project\McK water paper\codefiles\ERW_Mafic_Sept_05_2023.dat"

# Load the CSV file into a DataFrame
df = pd.read_csv(dfile_path)


# Initialize PHREEQC instance
phreeqc_mod.load_database(pqdb)  # Make sure you have PHREEQC database

results = []

for index, row in df.iterrows():
    # Create a PHREEQC input string for each row
    phreeqc_input = f"""
    SOLUTION 1
        units mmol/kgw
        pH 7 charge
        Ca {row['Ca']}
        Mg {row['Mg']}
        Na {row['Na']}
        Cl {row['Cl']}
    END
    RUN_CELLS
    """

    # Run PHREEQC
    phreeqc_mod.run_string(phreeqc_input)
    output = phreeqc_mod.get_selected_output_array()

    # Process output, e.g. get pH or other properties
    pH = output['pH'][0] if 'pH' in output.dtype.names else None
    results.append({'index': index, 'pH': pH})

# Convert results back to DataFrame
results_df = pd.DataFrame(results)
print(results_df)
```

```python
import pandas as pd
from phreeqpy.iphreeqc import IPhreeqc

# Paths
dfile_path = r"C:\Users\janel\OneDrive\Documents\IU grad\Research\Water-chem Project\McK water paper\codefiles\GLAQ.csv"
pqdb = r"C:\Users\janel\OneDrive\Documents\IU grad\Research\Water-chem Project\McK water paper\codefiles\ERW_Mafic_Sept_05_2023.dat"

# Load the CSV file into a DataFrame
df = pd.read_csv(dfile_path)

# Initialize PHREEQC instance
iphreeqc = IPhreeqc()
iphreeqc.load_database(pqdb)  # Use the variable, not a string 'pqdb'

results = []

for index, row in df.iterrows():
    phreeqc_input = f"""
    SOLUTION 1
        units mmol/kgw
        pH 7 charge
        Ca {row['Ca']}
        Mg {row['Mg']}
        Na {row['Na']}
        Cl {row['Cl']}
    END
    """

    iphreeqc.run_string(phreeqc_input)
    output = iphreeqc.get_selected_output_array()

    pH = output['pH'][0] if 'pH' in output.dtype.names else None
    results.append({'index': index, 'pH': pH})

results_df = pd.DataFrame(results)
print(results_df)
```
